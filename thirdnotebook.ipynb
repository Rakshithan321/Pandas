# Adding new values to Existing table

```python
# You can directly assign a list/series to a new column
```

```python
# create df
import pandas as pd
d={
    "Names":["Rakshi","Rashmi","Kiran","Punarv"],
    "Percentage":[82,70,75,95]
}
df1=pd.DataFrame(d)
df1
# Syntax:df["column_name"]=["values"]
# Add a new column "Grade"
df1["Grade"]=["A","B","B+","A+"]
df1
```

```python
# Syntax for adding single row: df.loc[index]=[----]
# Add new row at index 4
df1.loc[4]=["Arun",65,"C"]
df1
```

```python
# Syntax for multiple rows: pd.concat([df,new_df])
# example:
# create df
new_data=pd.DataFrame({
    "Names":["ramya","kavya"],
    "Percentage":[80,81],
    "Grade":["A","A"]
})
df2 = pd.concat([df1, new_data], ignore_index=True)
print(df2)
```

# Missing Value Imputation

```python
# In pandas, missing value imputation means filling in the NaN (Not a Number) or none values with some approxiate replacement
```

```python
d={
    "Names":["Hamsa","Varu","Punarv","Bhuvik"],
    "Age":[20,None,23,None],
    "Salary":[20000,45000,34000,50000]
}
df5=pd.DataFrame(d)
df5
```

```python
# count missing value per column
print(df5.isnull().sum())
```

```python
# Drop rows with any missing values
df5.dropna()
```

```python
# Drop column with missing values
df5.dropna(axis=1)
```

```python
# Replace all NaN with 0
df5.fillna(0)
```

```python
# Replace NaN in age column with 18
df5["Age"].fillna(18)
df5
```

```python
# Fill with column mean()
df5["Age"].fillna(df5["Age"].mean())
df5
```

```python
# Fill with column median()
df5["Salary"].fillna(df5["Salary"].median())
```

# Forward Fill And Backward Fill (Propagate Values)

```python
# Forward Fill
# Takes the last known value and it forward to fill the next missing cell
# Think of it as: "Repeat the previous value"
import pandas as pd
d={"A":[1,None,None,4,None,6]}
df6=pd.DataFrame(d)
print("Orginal Data:")
df6
```

```python
print("Forward Fill:")
print(df6.fillna(method="ffill"))
```

```python
print("\n Backward Fill:")
print(df6.fillna(method="bfill"))
```

# Map And Replace

```python
# In pandas,both map() and replace() are used totransform or modify values in a series or Dataframe column.But they are slightly different
```

```python
# Map()
# works only on a series (one cloumn)
# used for elemnts-wise transformations
# accepts a function,dictionary or series for mapping
import pandas as pd
# Sample Data
d={
    "Name":["Alice","Bob","Charlie","David"],
    "Gender":["F","M","M","F"]
}
df6=pd.DataFrame(d)
df6
```

```python
# Using dictionary with map()
df6["Gender"]=df6["Gender"].map({"M":"Male","F":"Female"})
df6
```

```python
# Replace()
# works on series and dataframe (multiple column)
# more flexible than map()
# can replace single value,multiple values, or even use regex
```

```python
# Replace values in column
df6["Name"]=df6["Name"].replace({"Alice":"Alicia","Bob":"Boddy"})
df6
```

# Merge

```python
# works like a SQL joins (inner,left,right,outer)
# Flexible: you can join on one or more columns
# Combines DataFrame based on common columns or indexes
```

```python
import pandas as pd
df9=pd.DataFrame({
        "Id":[1,2,3],
         "Names":["A","B","C"]
})
df10=pd.DataFrame({
    "Id":[1,2,4],
    "Score":[90,80,70]
})
# Merge
merged=pd.merge(df9,df10, on="Id", how="inner")
merged
```

```python
merged=pd.merge(df9,df10, on="Id", how="outer")
merged
```

```python
merged=pd.merge(df9,df10, on="Id", how="left")
merged
```

```python
merged=pd.merge(df9,df10, on="Id", how="right")
merged
```

# JOIN

```python
df9=pd.DataFrame({"Names":["A","B","C"]}, index=[1,2,3])
df10=pd.DataFrame({"Score":[90,80,70]}, index=[1,2,4])
joined=df9.join(df10, how="inner")
joined
```

# Conditional Filtering

```python
# It means selecting rows from a dataset (DataFrame) based on a condition (or multiple condition)
```

```python
import pandas as pd 
d={"name":["a","b","c","d","e"],"per":[89,56,78,45,60]}
df11=pd.DataFrame(d)
df11
```

```python
# Filtering with single codition
df11[df11.per<70]
```

```python
# Filtering with Multiple condition
# you can combine cindition with:
# &---> AND
# |---> OR
# ~---> NOT
```

```python
df11[(df11.per>60) & (df11.per<80)]
```

```python
# using query
df11.query("per>50")
```
